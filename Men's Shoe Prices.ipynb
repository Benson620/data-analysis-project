```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_theme(context="talk")
```

```python
df = pd.read_csv("file:///Users/linhongteng/Downloads/train.csv")
df.info()
```

```python
df.head()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
# calculate each column's missing value rate
round(df.isna().sum() / df.shape[0], 3)
```

```python
# pick the missing rate >= 80%, and drop them
cols_to_drop = [column for column in df.columns if df[column].isna().sum() >= 0.8*df.shape[0]]
cols_to_drop
```

```python
df.drop(columns=cols_to_drop, inplace=True)
df.columns
```

```python
# now, the price type is still object, need to change the type

df.prices_amountmin = pd.to_numeric(df.prices_amountmin, errors='coerce', downcast='float')
df.prices_amountmax = pd.to_numeric(df.prices_amountmax, errors='coerce', downcast='float')
df.info()
```

```python
# narrow the data and focus on some columns
df.drop(columns=['id', 'dateadded', 'dateupdated', 'descriptions', 'ean', 
                 'features', 'imageurls', 'keys', 'manufacturer',
                 'manufacturernumber', 'merchants', 'prices_condition',
                 'prices_dateadded', 'prices_dateseen', 'prices_issale',
                 'prices_merchant', 'prices_offer', 'prices_shipping',
                 'prices_sourceurls', 'sizes', 'skus', 'sourceurls', 'upc'],
                 inplace=True)
df.info()
```

```python
df.head()
```

```python
# find two pretty similar columns,"prices_amountmin" & "prices_amountmax", and calculate the corr rate
sum(df.prices_amountmin == df.prices_amountmax) / df.shape[0]
```

```python
df["price"] = np.mean([df.prices_amountmin, df.prices_amountmax], axis=0)
df_cleaned = df.drop(columns=["prices_amountmin", "prices_amountmax"])
df_cleaned.head()
```

```python
# there are more than one currency, remove another currency except "USD"
df_cleaned.prices_currency.value_counts()
```

```python
df_cleaned = df_cleaned[df_cleaned.prices_currency == "USD"]
df_cleaned.prices_currency.value_counts()
```

```python
# missing values
df_cleaned.isna().sum()
```

```python
df_cleaned.dropna(axis=0, subset=["brand"], inplace=True)
df_cleaned.isna().sum()
```

```python
# standardize brand's name
df_cleaned.brand = df_cleaned.brand.apply(lambda x: x.lower(), convert_dtype=False)
df_cleaned.brand.head()
```

```python
# total unique brands
df_cleaned.brand.nunique()
```

```python
# final cleaning data
df_cleaned.head(10)
```

```python
# find the brand.value_counts > 100
brand_above_100 = df_cleaned.brand.value_counts()[df_cleaned.brand.value_counts() > 100]
brand_above_100
```

```python
print("Brands of selling 100+ times list:\n", brand_above_100.index)
print("Total brands of selling 100+ times:\n", brand_above_100.count())
```

```python
df_brands = df_cleaned[df_cleaned.brand.isin(brand_above_100.index)].groupby("brand").mean()
df_brands.sort_values("price", ascending=False, inplace=True)
df_brands
```

```python
# bar chart: Average price by brands
fig = plt.figure(figsize=(10,12))
sns.barplot(data=df_brands, x="price", y=df_brands.index, palette="crest")
plt.title("Average price by brand")
plt.xlabel("Price")
plt.ylabel("Brand")
plt.annotate(text="Note: only brands with > 100 observations are included.", xy=(280,23), fontsize=10)
```

```python
# find which brand launched the highest price
df_max_prices = df_cleaned[df_cleaned.brand.isin(brand_above_100.index)].sort_values("price", ascending=False)
df_max_prices.head()
```

```python
df_max_prices = df_cleaned[df_cleaned.brand.isin(brand_above_100.index)].sort_values("price", ascending=False)
df_max_prices = df_max_prices.drop_duplicates("brand")
df_max_prices[["brand", "name", "price"]]
```

```python
# Max price by brands
fig = plt.figure(figsize=(10,12))
sns.barplot(data=df_max_prices, x="price", y="brand", palette="crest")
plt.title("Max price by brand")
plt.xlabel("Price")
plt.ylabel('Brand')
plt.annotate(text="Note: only brand with > 50 observations are included.", xy=(1400,23), fontsize=10)
```

```python
# which brand have the widest distribution of prices
df_medians = df_cleaned[df_cleaned.brand.isin(brand_above_100.index)]
df_medians = df_medians.groupby("brand").median()
df_medians = df_medians.sort_values("price", ascending=False)
df_medians
```

```python
# boxplots
fig=plt.figure(figsize=(8,16))
sns.boxplot(data=df_cleaned[df_cleaned.brand.isin(brand_above_100.index)],
            x="price",
            y="brand",
            palette="crest",
            order = df_medians.index,
            orient="h")

plt.title("Boxplots for price by brand")
plt.xlabel("Price")
plt.ylabel("Brand")
plt.annotate(text="Note: only brand with > 50 observations are included.", xy=(1400,23), fontsize=10)
```
